```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np 
import cv2 
```

```python
def color_threshold(img, thres=(0, 0, 0)):
    binary = np.zeros_like(img[:, :, 0])
    threshold = (img[:, :, 0] > thres[0]) & (img[:, :, 1] > thres[1]) & (img[:, :, 2] > thres[2])
    binary[threshold] = 1
    return binary

def perspective_transform(img, src, dst):
    M = cv2.getPerspectiveTransform(src, dst)
    wraped = cv2.warpPerspective(img, M, (image.shape[1], image.shape[0]))
    return wraped
```

```python

image = mpimg.imread("sample.jpg")
plt.imshow(image)
```

```python
src = np.float32([[14, 140], [300, 140], [120, 95], [200,95]])
dst = np.float32([[145, 140], [155, 140], [145, 130], [155, 130]])
```

```python
binary = color_threshold(image, thres=(180, 180, 180))
plt.imshow(binary, cmap='gray')
```

```python
warped = perspective_transform(binary, src, dst)
plt.imshow(warped, cmap='gray')
```

```python
def rover_coords(binary_img):
    nonzero_y, nonzero_x = binary_img.nonzero()
    
    pixel_x = -(nonzero_y - binary_img.shape[0]).astype(np.float32)
    pixel_y = -(nonzero_x - binary_img.shape[1]/2).astype(np.float32)
    return pixel_x, pixel_y
    
    
```

```python
pixel_x, pixel_y = rover_coords(warped)
plt.plot(pixel_x, pixel_y, '.')
plt.xlim(0, 360)
plt.ylim(150, -150)
plt.show()
```
