```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 
import numpy as np 
```

```python
image = mpimg.imread("example_grid1.jpg")
plt.imshow(image)
```

```python
def perspective_transform(img, src, dst):
    M = cv2.getPerspectiveTransform(src, dst)
    wraped_img = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    return wraped_img
```

```python
src = np.float32([[14, 140], [300, 140], [120, 95], [200,95]])
dst = np.float32([[145, 140], [155, 140], [145, 130], [155, 130]])
print(src[0, :])
```

```python
plt.scatter(src[0, 0], src[0, 1])
plt.scatter(src[1, 0], src[1, 1])
plt.scatter(src[2, 0], src[2, 1])
plt.scatter(src[3, 0], src[3, 1])
```

```python

dst_size = 5 
# Set a bottom offset to account for the fact that the bottom of the image 
# is not the position of the rover but a bit in front of it
bottom_offset = 6

destination = np.float32([[image.shape[1]/2 - dst_size, image.shape[0] - bottom_offset],
                  [image.shape[1]/2 + dst_size, image.shape[0] - bottom_offset],
                  [image.shape[1]/2 + dst_size, image.shape[0] - 2*dst_size - bottom_offset], 
                  [image.shape[1]/2 - dst_size, image.shape[0] - 2*dst_size - bottom_offset],
                  ])

source = np.float32([[14, 140], [301 ,140],[200, 96], [118, 96]])
wraped = perspective_transform(image, source, destination)
plt.imshow(wraped)
```

```python
wraped = perspective_transform(image, src, dst)
plt.imshow(wraped)
```
